```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
from langchain_huggingface import HuggingFaceEndpoint,ChatHuggingFace
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
llm = HuggingFaceEndpoint(
    repo_id="meta-llama/Llama-3.1-8B-Instruct",
    task="text-generation",
    huggingfacehub_api_token=os.getenv('HUGGINGFACEHUB_API_TOKEN')
)

model = ChatHuggingFace(llm=llm)
```

```python
class BlogLLM(TypedDict):
    topic:str
    outline:str
    blog:str
```

```python
def generate_outline(state:BlogLLM)->BlogLLM:
    topic=state['topic']

    prompt=f'Generate a detailed outline for a blog post on the following topic : {topic}'

    res=model.invoke(prompt)
    state['outline']=res.content
    return state
```

```python
def generate_blog(state:BlogLLM)->BlogLLM:
    outline=state['outline']

    prompt=f'Generate a detailed blog post based on the following outline: {outline}'

    res=model.invoke(prompt)
    state['blog']=res.content
    return state
```

```python
graph=StateGraph(BlogLLM)

graph.add_node('generate_outline',generate_outline)
graph.add_node('generate_blog',generate_blog)

graph.add_edge(START,'generate_outline')
graph.add_edge('generate_outline','generate_blog')
graph.add_edge('generate_blog',END)

workflow=graph.compile()

final_state=workflow.invoke({
    'topic': 'How to use Langgraph?'
})
```
